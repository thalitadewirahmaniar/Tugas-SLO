```python
#10 player terbanyak mencetak gol

# Mengelompokkan data berdasarkan pemain dan menghitung jumlah gol untuk setiap pemain
player_goals = data_noid.groupby('player_name')['goals'].sum()

# Mengurutkan pemain berdasarkan jumlah gol secara menurun
players_sort = player_goals.sort_values(ascending=False)

# Mengambil 10 pemain pertama setelah diurutkan
players_top10 = players_sort.head(10)

print(players_top10)
```

```python
# 2. 10 player terlama durasi bermain

# Mengelompokkan data berdasarkan pemain dan menjumlahkan durasi bermain untuk setiap pemain
player_time = data_noid.groupby('player_name')['time'].sum()

# Mengurutkan pemain berdasarkan total durasi bermain secara menurun
players_sort = player_time.sort_values(ascending=False)

# Mengambil 10 pemain pertama (top 10) setelah diurutkan
players_top10 = players_sort.head(10)

print(players_top10)
```

```python
# 3. 10 player terbanyak assist

# Mengelompokkan data berdasarkan pemain dan menjumlahkan assists untuk setiap pemain
player_assists = data_noid.groupby('player_name')['assists'].sum()

# Mengurutkan pemain berdasarkan total assists secara menurun
players_sort = player_assists.sort_values(ascending=False)

# Mengambil 10 pemain pertama (top 10) setelah diurutkan
players_top10 = players_sort.head(10)

print(players_top10)
```

```python
# 4 player pencetak gol terbanyak dari kesebelasan manchester city

# Filter baris berdasarkan kesebelasan Manchester City
mancity_players = data_noid[data_noid['team_title'] == 'Manchester City']

# Urutkan pemain-pemain tersebut berdasarkan jumlah gol secara menurun
mancity_players_sort = mancity_players.sort_values(by='goals', ascending=False)

# Ambil 4 pemain teratas
top4_mancity_players = mancity_players_sort.head(4)

print(top4_mancity_players[['player_name', 'goals']])
```

```python
# 5. dari kesebelasan pencetak gol terbanyak

# Mengelompokkan data berdasarkan kesebelasan dan menjumlahkan jumlah gol untuk setiap kesebelasan
teams_goals = data_noid.groupby('team_title')['goals'].sum()

# Mengurutkan kesebelasan berdasarkan jumlah gol secara menurun
teams_sort =teams_goals.sort_values(ascending=False)

# Mendapatkan kesebelasan dengan pencetak gol terbanyak
team_with_most_goals = teams_sort.index[0]

print("Kesebelasan dengan pencetak gol terbanyak :", team_with_most_goals)
print("Jumlah gol:", teams_goals[team_with_most_goals])
```

```python
# 6. Berapa jumlah kartu kuning dan merah yang diterima pembuat assist  terbanyak ?

# Temukan pembuat assist terbanyak
top_assist_maker = data_noid.groupby('player_name')['assists'].sum().idxmax()

# Filter data untuk pembuat assist terbanyak
top_assist_maker_data = data_noid[data_noid['player_name'] == top_assist_maker]

# Hitung jumlah kartu kuning dan merah yang diterima
yellow_cards = top_assist_maker_data['yellow_cards'].sum()
red_cards = top_assist_maker_data['red_cards'].sum()


print("Pembuat assist terbanyak:", top_assist_maker)
print("Jumlah kartu kuning yang diterima:", yellow_cards)
print("Jumlah kartu merah yang diterima:", red_cards)
```

```python
# 7. Siapa yang jumlah assist dan golnya paling banyak dan dari kesebelasan mana ?

# Mengelompokkan data berdasarkan pemain dan menjumlahkan assist dan gol
player_assists_goals = data_noid.groupby('player_name').agg({'assists': 'sum', 'goals': 'sum'})

# Temukan pemain dengan jumlah assist dan gol terbanyak
top_player = player_assists_goals[(player_assists_goals['assists'] == player_assists_goals['assists'].max()) & (player_assists_goals['goals'] == player_assists_goals['goals'].max())]

# Ambil kesebelasan asal pemain tersebut
team_title = data_noid[data_noid['player_name'] == top_player.index[0]]['team_title'].iloc[0]

# Menampilkan hasil
print("Pemain dengan jumlah assist dan gol terbanyak:", top_player.index[0])
print("Jumlah assist:", top_player['assists'].iloc[0])
print("Jumlah gol:", top_player['goals'].iloc[0])
print("Kesebelasan:", team_title)
```

```python
# 8. 5 kesebelasan yang menyumbangkan gol terbanyak

# Mengelompokkan data berdasarkan kesebelasan dan menjumlahkan jumlah gol
team_goals = data_noid.groupby('team_title')['goals'].sum()

# urut kesebelasan berdasarkan jumlah gol secara menurun
teams_sort = team_goals.sort_values(ascending=False)

# Ambil 5 kesebelasan teratas
top5_teams = teams_sort.head(5)

# Menampilkan hasil
print("5 kesebelasan yang menyumbangkan gol terbanyak:")
print(top5_teams)
```

```python
#9. Urutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali !

# Filter pemain yang tidak mendapatkan kartu kuning atau merah
players_nocards = data_noid[(data_noid['yellow_cards'] == 0) & (data_noid['red_cards'] == 0)]

# Urutkan pemain berdasarkan jumlah gol secara menurun
players_sort = players_nocards.sort_values(by='goals', ascending=False)

# Menampilkan hasil
print("Pemain dengan jumlah gol terbanyak tanpa kartu kuning atau merah:")
print(players_sort[['player_name', 'goals']])
```
