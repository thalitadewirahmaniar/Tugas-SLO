```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Membaca data dari CSV ke DataFrame
df = pd.read_csv("/content/drive/MyDrive/SLO/data0-SZL.csv", header=None)

df.describe()
```

```python
mean_values = df.mean()

std_values = df.std()

skewness = df.skew()

median = df.median()

energy = (df ** 2).sum()

min = df.min()

max = df.max()


df_data = pd.DataFrame({
    'Mean': mean_values,
    'Standard Deviation': std_values,
    'Skewness': skewness,
    'Median': median,
    'Energy': energy,
    'Min': min,
    'Max': max
})
df_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Pilih 6 kolom secara bebas untuk diplot
selected_columns = df.sample(6, axis=1)

# Plot sinyal dengan multiplot 2x3
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

for i, ax in enumerate(axes.flatten()):
    col = selected_columns.columns[i]
    ax.plot(selected_columns[col])
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(10, 6))

# Mengatur warna dan marker untuk setiap kelompok data 10 kolom
for i in range(0, len(mean_values), 10):
    if i // 10 % 16 == 0:
        color = 'red'
        marker = '*'
    elif i // 10 % 16 == 1:
        color = 'blue'
        marker = 'o'
    elif i // 10 % 16 == 2:
        color = 'green'
        marker = 'v'
    else:
        color = plt.cm.tab20(i // 10 % 20)  # Menggunakan colormap untuk mendapatkan warna yang berbeda
        marker = '.'
    plt.scatter(mean_values[i:i+10], std_values[i:i+10], label=f'Data Points {i+1}-{i+10}', color=color, marker=marker)

plt.xlabel('Mean')
plt.ylabel('Standard Deviation')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Scatter Plot')

plt.show()
```

```python

plt.figure(figsize=(10, 6))

# Mengatur warna dan marker untuk setiap kelompok data 10 kolom
for i in range(0, len(skewness), 10):
    if i // 10 % 16 == 0:
        color = 'red'
        marker = '*'
    elif i // 10 % 16 == 1:
        color = 'blue'
        marker = 'o'
    elif i // 10 % 16 == 2:
        color = 'green'
        marker = 'v'
    else:
        color = plt.cm.tab20(i // 10 % 20)  # Menggunakan colormap untuk mendapatkan warna yang berbeda
        marker = '.'
    plt.scatter(skewness[i:i+10], median[i:i+10], label=f'Data Points {i+1}-{i+10}', color=color, marker=marker)

plt.xlabel('Skewness')
plt.ylabel('Median')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Scatter Plot')

plt.show()
```
